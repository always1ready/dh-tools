Для качественного анализа нам нужно использовать **регулярные выражения (с помощью продвинутой библиотеки RegEx)**, чтобы различать «хорошие» дефисы (в словах *кто-то*, *социально-экономический*) и «плохие» (повторы типа *да-да*).

Существуют два модуля для работы с регулярными выражениями: встроенный `re` и сторонний `regex` (который часто называют `re2` или «продвинутый re»).

```python
import regex as re
```

**ЧАЩЕ ВСЕГО РАБОТАТЬ ВЫ БУДЕТЕ С ДВУМЯ МЕТОДАМИ:**
- `re.findall()` — ищет в тексте все фрагменты, которые подходят под описание, и возвращает их в виде списка

- `re.sub()` — ищет паттерн и заменяет его на другой (в том числе на пустую строку)

**ХОРОШО ЗНАТЬ И ДРУГИЕ:**
- `re.search()` — ищет первое совпадение в строке по заданному паттерну
- `re.match()` — ищет паттерн только по началу строки
- `re.fullmatch()` — ищет совпадение со всем текстом
- `re.finditer()` — возвращает все совпадения в виде итератора

**Регулярное выражение** — это строка, паттерн для поиска, но со спецсимволами.

```python
text = "Алиса-то не так уж удивилась, даже когда 5 раз услыхала, что Кролик сказал (а сказал он 'ай-ай-ай! уже 11:10! Я опаздываю!').\n"
```

```python
# СПЕЦЗНАКИ С БЭКСЛЕШЕМ
r"\w" # любая буква (или цифра)
r"\d" # любая цифра
r"\s" # любой пробельный символ (пробел, табуляция, перенос строки)

r"\W" # любая не буква
r"\D" # любая не цифра
r"\S" # любой непробельный символ (пробел, табуляция, перенос строки)

r'.' # любой символ

re.findall(r'\d', text)
```

Квадратные скобки используются для создания классов, карет (^) и знак доллара ($) – это якоря, но карет внутри квадратных скобок может также быть отрицанием. Точка – это любой знак, “джокер”.

```python
# выведите все пробельные символы

re.findall(r' ___ ', text)
```

```python
# КВАНТОРЫ
"а*" # ноль или более раз повторяется "а"
"a?" # ноль или один раз повторяется "а"
"a+" # один или более раз повторяется "а"

# затятые с пробелами
print(re.findall(r', ', text))
```

```python
# выведите количество комбинаций буква-дефис-буква

re.findall(r' ___ ', text)
```

```python
# ТОЧНОЕ КОЛИЧЕСТВО
"а{3,5}" # от 3 до 5 раз повторяется "а"
"a{5}"   # 5 раз повторяется "а"
"a{5,}"  # не менее 5 раз повторяется "а"
"a{,5}"  # не более 5 раз повторяется "а"
```

По умолчанию квантификаторам соответствует максимально длинная строка из возможных: квантификаторы являются жадными (greedy). Чтобы этого избежать, надо поставить после квантификатора знак вопроса. Это сделает его ленивым.

```python
"??"	    # 0 или 1, лучше 0
"*?"	    # 0 или больше, как можно меньше
"+?"	    # 1 или больше, как можно меньше
"{n,m}?"	# от n до m, как можно меньше
```

```python
# выведите встречаемость строки 'ай-'

re.findall(r' ___ ', text)
```

```python
# АЛЬТЕРНАТИВЫ
"abc|def" # либо "аbc", либо "def"
```

```python
# выведите список строк 'ай-' или 'ай'

re.findall(r' ___ ', text)
```

```python
# КВАДРАТНЫЕ СКОБКИ
"[abc]" # любой из символов: либо "а", либо "b", либо "с"  ЧУВСТВИТЕЛЬНО К РЕГИСТРУ
"a|b|c" # то же самое, но длиннее
"[a-z]" # любой из символов в диапазоне от "а" до "z"
"[^abc]" # любой символ, кроме "а", "b", "с"    -- знак "^" означает исключение
```

```python
# выведите список всех заглавных букв

re.findall(r' ___ ', text)
```

```python
# СПЕЦСИМВОЛЫ ДЛЯ ОТРЕЗКОВ НУЛЕВОЙ ДЛИНЫ
"^" # символ начала строки НЕ ПУТАТЬ С ГАЛОЧКОЙ В КВАДРАТНЫХ СКОБКАХ
"$" # символ конца строки
"\b" # между буквой (или цифрой) и не буквой (или цифрой)
```

#### Спецсимволы и пробельные последовательности
При работе с сырыми данными из интернета или распознанных PDF-файлов мы часто сталкиваемся с невидимыми символами, которые влияют на обработку текста. Регулярные выражения позволяют легко их находить.

Основные символы:

`\n` — перенос строки (newline) разделяет текст на абзацы.

`\t` — табуляция (отступ в начале строки).

`\r` — возврат каретки (часто встречается в файлах, созданных в Windows, в паре с `\n`).

По умолчанию символ точки `.` находит любой символ, кроме переноса строки `\n`

```python
re.findall(r'\n', text)
```

```python
# 1. re.findall() — найти все совпадения. 
# Поищем все группы цифр (например, годы или время)
# \d — любая цифра, + — одна или более
numbers = re.findall(r'\d+', text)
print(f"Цифры в тексте: {numbers}")

# 2. Поиск имен собственных
# [А-Я] — любая заглавная русская буква, [а-я]+ — одна или более строчных
names = re.findall(r'[А-Я][а-я]+', text)
print(f"Имена: {names}")

# 3. Шаблон: цифры-двоеточие-цифры
time_pattern = re.findall(r'\d{1,2}:\d{2}', text)
print(f"Время: {time_pattern}")
```

```python
# 2. re.sub() — заменить все совпадения, принимает уже два паттерна 1. что найти и 2. что заменить
# Заменим все группы цифр что-то одно
res1 = re.sub(r'\d+', r'505', text)
print(res1)
```

```python
# В АВТОЗАМЕНЕ СКОБКИ ОЗНАЧАЮТ ГРУППЫ

"\0" # всё выражение
"\1" # первая группа

re.sub(r"(\w+) (\w+)", r"\2 \1", "мама мыла")
```

Круглые скобки `()` создают группу захвата. Это значит, что программа запоминает то, что попало в скобки. 

При использовании функции замены `re.sub()` мы можем обратиться к этой сохраненной группе по её номеру (например, `\1` для первой группы).

```python
# Заменим все имена собственные, напишем их задом наперёд (а заодно повторим лямбда-функции)
res2 = re.sub(r'([А-Я][а-яё]*)', lambda m: m.group(0)[::-1], res1)
print(res2)
```

```python
# Добавим троеточие после всех слов, заканчивающихся на гласную
res3 = re.sub(r'([ауеяюоиэ])(\s)', r'\1... ', res2)
print(res3)
```

Сложные регулярные выражения трудно читать. Флаг `re.VERBOSE` (или `re.X`) позволяет разбить паттерн на несколько строк, игнорировать пробелы в самом паттерне и писать комментарии через `#`.

```python
pattern = r"""
    \b\w+\b   # Ищем одну и более букв со знаками границы слова
    ,          # затем следует запятая
    \s+         # а потом один или несколько пробелов
"""

# Обязательно передаем флаг re.VERBOSE третьим аргументом!
re.findall(pattern, text, re.VERBOSE)
```
